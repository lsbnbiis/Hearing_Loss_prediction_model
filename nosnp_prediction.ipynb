```python
import pandas as pd
import numpy as np
```

```python
def ectract_male_cate_nosnp(input_cate_df):
    re_cate_df = input_cate_df.filter(items=['JOB_LGST_OCCUPATION', 'DRK', 'SMK_EXPERIENCE', 'SPO_HABIT',
       'ORTHOPEDICS_ARTICULUS', 'OSTEOPOROSIS', 'ARTHRITIS',
       'HEART_BLOOD_TUBE', 'HYPERLIPIDEMIA', 'HYPERTENSION', 'APOPLEXIA',
       'DIABETES', 'HEMICRANIA', 'VERTIGO', 'CARCINOMA_IN_SITU',
       'ARTICULUS_ACHE', 'NECK_ACHE', 'BACK_AND_WAIST_ACHE', 'SICATICA',
       'HEADACHE_AND_HEMICRANIA', 'I_4', 'I_5', 'I_7', 'I_11', 'I_12', 'I_13',
       'I_14', 'I_15', 'I_16', 'I_17', 'I_18', 'I_21', 'I_22', 'I_23', 'I_24',
       'I_25', 'I_28', 'I_31', 'HBA1C', 'ANTI_HCV_AB_1', 'HBSAG_1', 'HBEAG_1',
       'ANTI_HBS_AB_1', 'ANTI_HBC_AB_1'], axis=1)
    return re_cate_df

def ectract_male_num_nosnp(input_num_df):
    re_num_df = input_num_df.filter(items=['AGE', 'BODY_HEIGHT', 'BODY_WEIGHT', 'BMI', 'BODY_WAISTLINE',
       'BODY_BUTTOCKS', 'WHR', 'T_SCORE', 'BODY_FAT_RATE', 'RBC', 'WBC',
       'PLATELET', 'HB', 'HCT', 'FASTING_GLUCOSE', 'T_CHO', 'TG', 'HDL_C',
       'LDL_C', 'T_BILIRUBIN', 'ALBUMIN', 'SGOT', 'SGPT', 'GAMMA_GT', 'AFP',
       'BUN', 'CREATININE', 'URIC_ACID', 'MICROALB',
       'SIT_SYSTOLIC_PRESSURE_AVERAGE', 'SIT_DIASTOLIC_PRESSURE_AVERAGE',
       'SIT_HEARTBEAT_SPEED_AVERAGE'], axis=1) 
    return re_num_df
    
def ectract_female_cate_nosnp(input_cate_df):
    re_cate_df = input_cate_df.filter(items=['JOB_LGST_OCCUPATION', 'DRK', 'SMK_EXPERIENCE', 'SPO_HABIT',
       'ORTHOPEDICS_ARTICULUS', 'OSTEOPOROSIS', 'ARTHRITIS',
       'HEART_BLOOD_TUBE', 'HYPERLIPIDEMIA', 'HYPERTENSION', 'APOPLEXIA',
       'DIABETES', 'HEMICRANIA', 'VERTIGO', 'CARCINOMA_IN_SITU',
       'ARTICULUS_ACHE', 'NECK_ACHE', 'BACK_AND_WAIST_ACHE', 'SICATICA',
       'HEADACHE_AND_HEMICRANIA', 'MENSTRURATION_REGULAR',
       'MENSTRURATION_CURR', 'SEVERE_DYSMENORRHEA', 'MYOMA', 'OVARIAN_CYST',
       'ENDOMETRIOSIS', 'CERVICAL_POLYP', 'UTERINE_CA', 'CERVICAL_CA',
       'OVARIAN_CA', 'I_4', 'I_5', 'I_7', 'I_11', 'I_12', 'I_13', 'I_14',
       'I_15', 'I_16', 'I_17', 'I_18', 'I_21', 'I_22', 'I_23', 'I_24', 'I_25',
       'I_28', 'I_31', 'HBA1C', 'ANTI_HCV_AB_1', 'HBSAG_1', 'HBEAG_1',
       'ANTI_HBS_AB_1', 'ANTI_HBC_AB_1'], axis=1)
    return re_cate_df

def ectract_female_num_nosnp(input_num_df):
    re_num_df = input_num_df.filter(items=['AGE', 'MENARCHE', 'BODY_HEIGHT', 'BODY_WEIGHT', 'BMI',
       'BODY_WAISTLINE', 'BODY_BUTTOCKS', 'WHR', 'T_SCORE', 'BODY_FAT_RATE',
       'RBC', 'WBC', 'PLATELET', 'HB', 'HCT', 'FASTING_GLUCOSE', 'T_CHO', 'TG',
       'HDL_C', 'LDL_C', 'T_BILIRUBIN', 'ALBUMIN', 'SGOT', 'SGPT', 'GAMMA_GT',
       'AFP', 'BUN', 'CREATININE', 'URIC_ACID', 'MICROALB', 'MC_DURATION',
       'SIT_SYSTOLIC_PRESSURE_AVERAGE', 'SIT_DIASTOLIC_PRESSURE_AVERAGE',
       'SIT_HEARTBEAT_SPEED_AVERAGE'], axis=1)
    return re_num_df

from sklearn.preprocessing import StandardScaler
def generate_pred_format(input_cate, input_num):
    cate_array = np.array(input_cate, dtype='float32')
    num_array = np.array(input_num, dtype='float32')
    
    return np.hstack([cate_array,num_array])
```

```python
# Read example, scaledata and trans to numpy array
test_df = pd.read_csv("nosnp_test_data.csv", encoding="utf-8-sig")
male_test = test_df[(test_df["SEX"].astype(int) == 1)]
female_test = test_df[test_df["SEX"].astype(int) == 2]

cate_male_test_df = ectract_male_cate_nosnp(male_test)
num_male_test_df = ectract_male_num_nosnp(male_test)
cate_female_test_df = ectract_female_cate_nosnp(female_test)
num_female_test_df = ectract_female_num_nosnp(female_test)
```

```python
male_test_array = generate_pred_format(cate_male_test_df, num_male_test_df)
female_test_array = generate_pred_format(cate_female_test_df, num_female_test_df)
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

def draw_pic(r_data, res):
    labels = r_data.iloc[:,0]

    # 合併成 dataframe
    plt_df = pd.DataFrame({
        "Sample": labels,
        "pred_res": res.astype(int)
    })

    # 顏色
    colors = ["green" if v==0 else "red" for v in res]
    fig, ax = plt.subplots(figsize=(2,6))
    n = len(res)

    # 畫格子邊界
    for i in range(n):
        ax.add_patch(Rectangle((-0.5, i-0.5),1,1,
                           fill=False,
                           edgecolor='black',
                           linewidth=1))

    # 小色塊大小
    block_w = 0.4
    block_h = 0.4
    for i,v in enumerate(res.astype(int)):
        color = "red" if v==1 else "green"

        ax.add_patch(Rectangle((-block_w/2, i-block_h/2),
                           block_w,
                           block_h,
                           facecolor=color,
                           edgecolor='black',
                           linewidth=0.8))

    # 座標設定
    ax.set_xlim(-0.5,0.5)
    ax.set_ylim(-0.5,n-0.5)

    # y軸反轉
    ax.invert_yaxis()
    ax.set_xticks([0])
    ax.set_xticklabels(["Prediction\nResult"])

    # 把 X 軸移到上方
    ax.xaxis.tick_top()
    ax.xaxis.set_label_position('top')
    ax.set_yticks(np.arange(n))
    ax.set_yticklabels(labels)

    # 讓色塊為正方形
    ax.set_aspect('equal')
    plt.tight_layout()
    plt.show()
```

```python
#男性結果，0表示預測結果為無聽損，1為有聽損
import joblib

svm_model = joblib.load("nosnp_male_svm_model.joblib")
svm_score = svm_model.predict(male_test_array)

draw_pic(male_test, svm_score)
```

```python
#女性預測結果
import joblib

svm_model = joblib.load("nosnp_female_svm_model.joblib")
svm_score = svm_model.predict(female_test_array)

draw_pic(female_test, svm_score)
```
