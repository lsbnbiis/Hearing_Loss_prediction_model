```python
# Read example and trans to numpy array
test_df = pd.read_csv("snp_test_data.csv", encoding="utf-8-sig")
test_df["SEX"] = test_df["SEX"].astype(int)
male_test = test_df[(test_df['SEX'] == 1)]
female_test = test_df[test_df["SEX"] == 2]

cate_male_test_df = ectract_male_cate_snp(male_test)
num_male_test_df = ectract_male_num_snp(male_test)
cate_female_test_df = ectract_female_cate_snp(female_test)
num_female_test_df = ectract_female_num_snp(female_test)

cate_male_array = np.array(cate_male_test_df, dtype='float32')
num_male_array = np.array(num_male_test_df, dtype='float32')
cate_female_array = np.array(cate_female_test_df, dtype='float32')
num_female_array = np.array(num_female_test_df, dtype='float32')
```

```python
#數值資料正規化
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(num_female_array)
num_female_array = scaler.transform(num_female_array)
female_test_array = np.hstack([cate_female_array,num_female_array])

scaler = StandardScaler()
scaler.fit(num_male_array)
num_male_array = scaler.transform(num_male_array)
male_test_array = np.hstack([cate_male_array,num_male_array])
```

```python
#男性結果，0表示預測結果為無聽損，1為有聽損
import joblib

svm_model = joblib.load("tpmi_male_svm_model.joblib")
svm_score = svm_model.predict(male_test_array)
print("SVM:", svm_score)

rf_model = joblib.load("tpmi_male_rf_model.joblib")
rf_score = rf_model.predict(male_test_array)
print("RF:", rf_score)

xgboost_model = joblib.load("tpmi_male_xgboost_model.joblib")
xgboost_score = xgboost_model.predict(male_test_array)
print("XGBoost:", xgboost_score)

dtree_model = joblib.load("tpmi_male_dtree_model.joblib")
dtree_score = dtree_model.predict(male_test_array)
print("DecisionTree:", dtree_score)

HGBC_model = joblib.load("tpmi_male_HGBC_model.joblib")
HGBC_score = HGBC_model.predict(male_test_array)
print("HGBC:", HGBC_score)
```
